```python
import ujson
import pandas
```

```python
data = pandas.read_json('stargazers_paginated_jiahaog_nativefier.json')
contrib_data = pandas.read_json('nativefier_contrib.json')
```

```python
contrib_data

```

```python
contrib_data["contributor"]=contrib_data.apply(lambda x: x['author']['login'], axis=1)
```

```python
contrib_data
```

```python
cs = contrib_data[['weeks','contributor']].set_index('contributor')
```

```python
contrib_pivoted = cs.weeks.apply(pandas.Series).stack().to_frame().rename(columns={0:"weeks"})
```

```python
contrib_pivoted
```

```python
contrib_pivoted["time"]=contrib_pivoted.apply(lambda x: x['weeks']['w'], axis=1)
contrib_pivoted["added"]=contrib_pivoted.apply(lambda x: x['weeks']['a'], axis=1)
contrib_pivoted["deleted"]=contrib_pivoted.apply(lambda x: x['weeks']['d'], axis=1)
contrib_pivoted["commits"]=contrib_pivoted.apply(lambda x: x['weeks']['c'], axis=1)
```

```python
contrib_pivoted
```

```python
data.to_csv("test_stars.csv")
contrib_data.to_csv("test_natifier_data.csv")
contrib_pivoted.to_csv("natifier_data.csv")
```
